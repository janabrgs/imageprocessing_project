```python
import os
from ultralytics import YOLO

# === DEINE PFADE ANPASSEN (so wie bei dir) ===
PROJECT = r"C:\Users\edson\Downloads\Flashcard_Projekt"
DATA_YAML = os.path.join(PROJECT, "data.yaml")

# Trainierte Modelle (bei dir unter runs\detect\trainX\weights\best.pt)
MODEL_TRAIN2 = r"C:\Users\edson\runs\detect\train2\weights\best.pt"
MODEL_TRAIN3 = r"C:\Users\edson\runs\detect\train3\weights\best.pt"
MODEL_TRAIN4 = r"C:\Users\edson\runs\detect\train4\weights\best.pt"

print("DATA_YAML:", DATA_YAML)
print("exists data.yaml?", os.path.exists(DATA_YAML))
print("exists train4 best.pt?", os.path.exists(MODEL_TRAIN4))
```

```python
# START FROM PRETRAINED (so hast du angefangen)
model = YOLO("yolov8n.pt")

# Train (wie dein CLI: yolo detect train model=yolov8n.pt data=... epochs=... imgsz=...)
# ACHTUNG: Das erstellt runs/detect/train... neu, wenn du es ausführst.
model.train(
    data=DATA_YAML,
    epochs=30,
    imgsz=640,
    name="train_notebook"
)
```

```python
# Fine-tuning: Startet von einem existierenden best.pt (so wie bei dir train3/train4)
model = YOLO(MODEL_TRAIN3)  # oder MODEL_TRAIN2 / MODEL_TRAIN4

model.train(
    data=DATA_YAML,
    epochs=10,
    imgsz=640,
    name="train_finetune_notebook"
)
```

```python
# Nimm dein bestes Modell (z.B. train4)
model = YOLO(MODEL_TRAIN4)

# Beispiel: auf einen Ordner laufen lassen (images/val oder images)
SOURCE = os.path.join(PROJECT, "images", "val")  # oder "images" für alles

# conf=0.05 war bei dir entscheidend
results = model.predict(
    source=SOURCE,
    conf=0.05,
    save=True,       # speichert Bilder mit Boxen in runs/detect/predict...
    save_crop=True,  # speichert crops/card/...
    name="predict_notebook_conf005"
)

print("Fertig. Ergebnisordner ist unter: runs/detect/predict_notebook_conf005 (in deinem User-Ordner)")
```
